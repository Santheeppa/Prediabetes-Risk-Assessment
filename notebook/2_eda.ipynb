# 4.) Data Visualization

```python
def create_visualizations(df):
    # Age distribution by diabetes status
    plt.figure(figsize=(15, 5))
    
    plt.subplot(1, 3, 1)
    sns.boxplot(x='Prediabetes', y='Age', data=df, palette='Set2')
    plt.title('Age Distribution by Prediabetes Status')
    plt.xlabel('Prediabetes Status')
    plt.ylabel('Age')
    
    plt.subplot(1, 3, 2)
    sns.boxplot(x='Diabetes_Type', y='BMI', data=df, palette='Set2')
    plt.title('BMI Distribution by Diabetes Type')
    plt.xlabel('Diabetes Type')
    plt.ylabel('BMI')
    plt.xticks(rotation=45)
    
    plt.subplot(1, 3, 3)
    sns.boxplot(x='Physical_Activity_Level', y='HbA1c', data=df, palette='Set2')
    plt.title('HbA1c Levels by Physical Activity')
    plt.xlabel('Physical Activity Level')
    plt.ylabel('HbA1c')
    plt.xticks(rotation=45)
    
    plt.tight_layout()
    plt.show()
    
    # Additional visualizations
    fig, axes = plt.subplots(2, 2, figsize=(15, 12))
    
    # Region-wise diabetes distribution
    sns.countplot(data=df, x='Region', hue='Prediabetes', palette='Set2', ax=axes[0, 0])
    axes[0, 0].set_title('Region-wise Prediabetes Distribution')
    axes[0, 0].set_xlabel('Region')
    axes[0, 0].set_ylabel('Count')
    axes[0, 0].tick_params(axis='x', rotation=45)
    
    # BMI vs HbA1c with diabetes status
    sns.scatterplot(data=df, x='BMI', y='HbA1c', hue='Prediabetes', palette='Set2', ax=axes[0, 1])
    axes[0, 1].set_title('BMI vs HbA1c by Prediabetes Status')
    axes[0, 1].set_xlabel('BMI')
    axes[0, 1].set_ylabel('HbA1c')
    
    # Stress level distribution
    sns.boxplot(data=df, x='Prediabetes', y='Stress_Level', palette='Set2', ax=axes[1, 0])
    axes[1, 0].set_title('Stress Levels by Prediabetes Status')
    axes[1, 0].set_xlabel('Prediabetes Status')
    axes[1, 0].set_ylabel('Stress Level')
    
    # Screen time analysis
    sns.violinplot(data=df, x='Prediabetes', y='Screen_Time', palette='Set2', ax=axes[1, 1])
    axes[1, 1].set_title('Screen Time Distribution by Prediabetes Status')
    axes[1, 1].set_xlabel('Prediabetes Status')
    axes[1, 1].set_ylabel('Screen Time')
    
    plt.tight_layout()
    plt.show()

create_visualizations(df)
```

Based on the data visualizations created in the previous cells, here are some insights:

1. **Age Distribution by Prediabetes Status**:
    - The boxplot shows the distribution of ages for individuals with and without prediabetes. It helps to identify if age has any significant impact on prediabetes status.

2. **BMI Distribution by Diabetes Type**:
    - The boxplot illustrates the BMI distribution across different types of diabetes. This can help in understanding the relationship between BMI and the type of diabetes.

3. **HbA1c Levels by Physical Activity**:
    - The boxplot shows how HbA1c levels vary with different levels of physical activity. This can provide insights into how physical activity affects blood sugar control.

4. **Region-wise Prediabetes Distribution**:
    - The count plot shows the distribution of prediabetes status across different regions. This can help identify if certain regions have higher rates of prediabetes.

5. **BMI vs HbA1c by Prediabetes Status**:
    - The scatter plot shows the relationship between BMI and HbA1c levels, colored by prediabetes status. This can help identify any correlation between BMI and blood sugar levels.

6. **Stress Levels by Prediabetes Status**:
    - The boxplot shows the distribution of stress levels for individuals with and without prediabetes. This can help understand if stress levels are associated with prediabetes.

7. **Screen Time Distribution by Prediabetes Status**:
    - The violin plot shows the distribution of screen time for individuals with and without prediabetes. This can help identify if screen time is a factor in prediabetes.

These visualizations provide a comprehensive understanding of how various factors such as age, BMI, physical activity, region, stress levels, and screen time are associated with prediabetes and diabetes.
