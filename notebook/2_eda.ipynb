## 3.) Descriptive statistics of data

### Data types

### Statistics

```python
df.describe()
```


### Insights from Summary Statistics

1. **Age**: Individuals are aged between 15 to 25 years, with an average age of 20 years, focusing on young individuals.

2. **Family Income**: Family income varies greatly, averaging around 1.3 million INR, indicating significant income variability.

3. **Genetic Risk Score**: Scores range from 1 to 10, with an average of 5.5, suggesting a moderate genetic predisposition to diabetes.

4. **BMI**: BMI values range from 16 to 40, with an average of 28, indicating that individuals are generally overweight.

5. **Fast Food Intake**: Frequency ranges from 0 to 10, with an average of 5, showing common fast food consumption.

6. **Fasting Blood Sugar**: Levels range from 70 to 180 mg/dL, with an average of 125 mg/dL, indicating some individuals may have elevated blood sugar.

7. **HbA1c**: Levels range from 4% to 10%, with an average of 7%, suggesting some individuals may have prediabetes or diabetes.

8. **Cholesterol Level**: Levels range from 120 to 300 mg/dL, with an average of 210 mg/dL, indicating some individuals may have high cholesterol.

9. **Sleep Hours**: Sleep ranges from 4 to 10 hours, with an average of 7 hours, indicating reasonable sleep duration.

10. **Stress Level**: Stress levels range from 1 to 10, with an average of 5.5, indicating moderate stress.

11. **Screen Time**: Screen time ranges from 1 to 12 hours, with an average of 6.5 hours, indicating significant daily screen usage.

These insights help understand the health and lifestyle patterns of the individuals and identify potential prediabetes risk factors.

# 4.) Data Visualization

```python
# Ensure 'Prediabetes' is treated as a categorical variable
df['Prediabetes'] = df['Prediabetes'].astype('category')

# Plot the histogram
sns.histplot(data=df, x='Age', hue='Prediabetes')
```

```python
sns.histplot(data=df, x='Age', y= 'Gender',hue='Prediabetes',palette='Set1')
```

```python
sns.histplot(data=df, x='BMI', hue='Prediabetes')
```

```python
sns.histplot(data=df, x='Cholesterol_Level',kde= True, hue='Prediabetes')
```

```python
sns.countplot(data=df,x='Prediabetes')
```

# Count Plot

```python
plt.figure(figsize=(10,6),dpi=100)
sns.countplot(data=df,x='Dietary_Habits',hue='Prediabetes')
plt.title('Dietary Habits vs Prediabetes')
plt.show()
```

```python
plt.figure(figsize=(10,6),dpi=100)
sns.countplot(data=df,x='Gender',hue='Prediabetes')
plt.title("prediabetes on gender")
plt.show()
```

```python
plt.figure(figsize=(10,6),dpi=100)
sns.countplot(data=df,x='Age',hue='Prediabetes')
plt.title('Prediabetes on Age')
plt.show()
```

# Boxplot

```python
plt.figure(figsize=(10,6),dpi=100)
sns.boxplot(data=df,x='Age',y='Prediabetes')
plt.show()
```
